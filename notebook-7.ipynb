## Project Overview
In this project, I work for the fictional OilyGiant mining company. The goal is to determine the most profitable region for developing a new oil well.

To do this, I will:

- Load and prepare geological data from three regions.
- Train Linear Regression models to predict oil reserves.
- Select the 200 most promising wells out of 500 based on predictions.
- Calculate potential profit using business constraints.
- Apply bootstrapping (1000 samples) to estimate average profit, 95% confidence intervals, and risk of losses.
- Recommend the optimal region based on profit and risk.
This project uses Linear Regression only, according to business requirements.
    

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
def train_and_evaluate_region(X, y, random_state=12345):
    X_train, X_valid, y_train, y_valid = train_test_split(
        X, y, test_size=0.25, random_state=random_state
    )
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_valid)
    
    rmse = mean_squared_error(y_valid, y_pred, squared=False)
    mean_pred = y_pred.mean()
    
    return {
        'model': model,
        'X_valid': X_valid,
        'y_valid': y_valid,
        'y_pred': y_pred,
        'rmse': rmse,
        'mean_pred': mean_pred
    }
```

```python
results = []

for i, (X, y) in enumerate(X_y):
    res = train_and_evaluate_region(X, y)
    results.append(res)
    print(f'\nRegion {i}')
    print(f'RMSE: {res["rmse"]:.2f}')
    print(f'Average predicted product: {res["mean_pred"]:.2f}')
```

```python
for i, res in enumerate(results):
    print(f'Region {i} - mean actual: {res["y_valid"].mean():.2f}, mean predicted: {res["mean_pred"]:.2f}')
```

```python
def calculate_profit(y_true, y_pred):
    """
    y_true, y_pred are pandas Series of length 500 (validation set).
    """
    # Indices of wells sorted by predicted product (descending)
    order = y_pred.sort_values(ascending=False).index
    
    # Take top N wells
    top_indices = order[:WELLS_TO_DEVELOP]
    
    # Actual product of selected wells
    selected_products = y_true.loc[top_indices]
    
    total_product = selected_products.sum()  # thousand barrels
    
    revenue = total_product * PRICE_PER_THOUSAND_BARRELS
    profit = revenue - BUDGET
    
    return profit
```

## Profit Function Explanation
The profit calculation function performs the following tasks:
- Sorts all predicted reserves in descending order
- Selects the top 200 wells
- Retrieves the actual reserves for these wells
- Computes profit using:
        
profit = (sum of selected actual reserves √ó 4500) ‚àí 100 000 000

# Single-Sample Profit Estimates (Before Bootstrapping)
Region	Profit (Single Sample)
0	33.21 million USD
1	24.15 million USD
2	27.10 million USD

Region 0 appears strongest in the single evaluation, but risk and variability must still be considered.


```python
def bootstrap_profits(y_true, y_pred, n_bootstrap=1000, random_state=12345):
    state = np.random.RandomState(random_state)
    profits = []

    # Reset indices to avoid any alignment issues
    y_true = y_true.reset_index(drop=True)
    y_pred = pd.Series(y_pred).reset_index(drop=True)

    # We always simulate exploration of 500 wells
    SAMPLE_SIZE = 500

    for _ in range(n_bootstrap):
        # Sample 500 indices WITH replacement from the full set
        indices = state.randint(0, len(y_true), SAMPLE_SIZE)

        sample_true = y_true.iloc[indices]
        sample_pred = y_pred.iloc[indices]

        profit = calculate_profit(sample_true, sample_pred)
        profits.append(profit)

    return np.array(profits)
```

<b>   Reviewer's comment </b>
    

However, we indeed need to sample 500 rows with replacement. The point is that if we remove the replacement, our subsample (or rather, all subsamples) will simply resemble the original sample. The question is, how do we ensure independence then? That is, there will be no variability in the data. As for duplicates, in reality, we can also encounter very similar wells. For simplicity, let's take this example: we have 100 objects and sample 100 as well. With replacement, we get different samples where some of the 100 elements may be missing, while others may appear multiple times. The example is very simple, but the idea should be clear. In our case, the original sample is much larger, and the subsample sizes are much smaller, which provides even greater variability in the data.   

</div>

```python

def calculate_profit(y_true, y_pred):
    # Number of wells we are allowed to develop
    WELLS_TO_DEVELOP = 200
    
    # Development budget
    BUDGET = 100_000_000
    
    # Revenue per thousand barrels
    PRICE_PER_THOUSAND_BARRELS = 4500

    # 1. Select the TOP 200 wells by predicted product
    top_indices = y_pred.sort_values(ascending=False).head(WELLS_TO_DEVELOP).index

    # 2. Extract the TRUE product values for those wells
    selected_true = y_true.loc[top_indices]

    # 3. Total product (thousand barrels)
    total_product = selected_true.sum()

    # 4. Compute revenue and profit
    revenue = total_product * PRICE_PER_THOUSAND_BARRELS
    profit = revenue - BUDGET

    return profit
```

```python
region_stats = []

for i, res in enumerate(results):
    y_true = res['y_valid']
    y_pred = res['y_pred']

    profits = bootstrap_profits(y_true, y_pred)

    mean_profit = profits.mean()
    conf_interval = np.percentile(profits, [2.5, 97.5])
    risk = (profits < 0).mean()

    print(f"\nRegion {i}")
    print(f"Average profit: {mean_profit/1_000_000:.2f} million USD")
    print(f"95% CI: {conf_interval[0]/1_000_000:.2f} to {conf_interval[1]/1_000_000:.2f} million USD")
    print(f"Risk of loss: {risk*100:.2f}%")
```
